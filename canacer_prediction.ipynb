```python
X=df.iloc[:,2:31].values
Y=df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
X_train=StandardScaler().fit_transform(X_train)
X_test=StandardScaler().fit_transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming Y_train is your continuous target variable
label_encoder = LabelEncoder()
label_encoder.fit(Y_train)  # Ensure this line is present
label_encoder.fit(Y_test)
Y_train_encoded = label_encoder.fit_transform(Y_train)
# Encode the target variable for testing (using the same encoder)
Y_test_encoded = label_encoder.transform(Y_test)

# Now Y_train_encoded contains integer labels suitable for classification
# You can use these encoded labels to train your classifier
```

```python
# models/ Algorithms

def models(X_train,Y_train_encoded):
        #logistic regression
        from sklearn.linear_model import LogisticRegression
        log=LogisticRegression(random_state=0)
        log.fit(X_train,Y_train_encoded)
        
        
        #Decision Tree
        from sklearn.tree import DecisionTreeClassifier
        tree=DecisionTreeClassifier(random_state=0,criterion="entropy")
        tree.fit(X_train,Y_train_encoded)
        
        #Random Forest
        from sklearn.ensemble import RandomForestClassifier
        forest=RandomForestClassifier(random_state=0,criterion="entropy",n_estimators=10)
        forest.fit(X_train,Y_train_encoded)
        
        print('[0]logistic regression accuracy:',log.score(X_train,Y_train_encoded))
        print('[1]Decision tree accuracy:',tree.score(X_train,Y_train_encoded))
        print('[2]Random forest accuracy:',forest.score(X_train,Y_train_encoded))
        
        return log,tree,forest
```

```python
model=models(X_train,Y_train_encoded)
```

```python
print(Y_train_encoded[:10])
```

```python
# testing the models/result

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

for i in range(len(model)):
    print("Model",i)
    print(classification_report(Y_test_encoded,model[i].predict(X_test)))
    print('Accuracy : ',accuracy_score(Y_test_encoded,model[i].predict(X_test)))
```
